```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.1)
y = lambda x : np.sin(x)

plt.plot(x,y(x))
plt.plot(x,y(x)-3)
plt.plot(x,1/2*y(1/2*x))
```
